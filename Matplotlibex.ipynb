```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#line graph
```

```python
x=np.arange(1,11)
y=np.arange(20,220,20)
z=np.arange(3,300,30)
plt.plot(x,y,label="x-y data",color='red',marker='s',lw=3,mec='b',mew=2)    #Plot y versus x as lines and/or markers. lw=line width
# ls - line style -,--,-.,:,  ls; supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
# mec - markeredgecolor , mew= marker edge width ,ms=marker size
plt.plot(x,z,label="x-z data",color='#33FF4C',marker='1',ls='--',mec='r',mew=2,ms=10)    #    #rgb   0-f    000000-black ffffff- white   ff0000 - red  00ff00-green  0000ff-blue
plt.xlabel("x axis")  # label x axis
plt.ylabel("y axis")   # label y axis
plt.title("My Graph")   # give heading / title to graph
plt.legend(loc='best',ncol=2)  # ncol  is number of columns , loc is location of legend box
plt.xticks(x)
plt.yticks(z)
plt.show()  # Display all open figures.
```

```python
x=np.arange(-3.15,3.15,0.05)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,"r+",label="sin curve")
plt.plot(x,z,"g*",label="cos curve")
plt.legend()
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])  # rect : tuple (left, bottom, width, height)
x=np.arange(1,11)
y=np.arange(2,22,2)
ax.plot(x,y)
ax.set_title("Graph using axes")
ax.set_xlabel("x axis")
ax.set_ylabel("y axis")
ax.set_xticks(np.arange(0,21))
ax.set_yticks(np.arange(0,31))
ax.set_xlim((0,20))
ax.set_ylim((0,30))
plt.show()
```

```python
countries =["india","china","japan","pakistan"]
population=[1388163000,1411750000,124470000,220425254	]
fig =plt.figure()
ax=fig.add_axes((0,0,1,1))
ax.bar(countries, population,color=['orange','red','blue','green'],width=0.5)
plt.show()
```

```python
countries =["india","china","japan","pakistan"]
population=[1388163000,1411750000,124470000,220425254	]
fig =plt.figure()
ax=fig.add_axes((0,0,1,1))

ax.barh(countries, population,color=['orange','red','blue','green'])  # barh() horizontal graph
plt.show()
```

```python
oil=[50,40,30,70]
water=[50,60,70,80]
brand=['oil1','oil2','oil3','oil4']
fig=plt.figure()
ax=fig.add_axes((0,0,1,1))
ax.bar(brand,water,color='b',width=0.40)
ax.bar(brand,oil,color='y',width=0.40,bottom=water)
ax.set_yticks(np.arange(10,160,10))
ax.legend(labels=["Water","Oil"])
plt.show()
```

```python
x=np.arange(1,11)
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(8,4))
ax[0][0].plot(x,x**2,'r',ls='--',lw=3)
ax[0][0].set_xlabel(" x value")
ax[0][0].set_ylabel("square of x")
ax[0][0].grid(True)
ax[0][1].plot(x,np.log(x),'g',ls='--',lw=3)
ax[0][1].set_xlabel(" x value")
ax[0][1].set_ylabel("log of x")
ax[1][0].plot(x,np.tanh(x),'b',ls=':',lw=1)
ax[1][0].set_xlabel(" x value")
ax[1][0].set_ylabel("tanh of x")
ax[1][1].plot(x,x,'b',ls='-',lw=2)
ax[1][1].set_xlabel(" x value")
ax[1][1].set_ylabel(" x")
ax[1][1].grid(color='r',ls=':',lw=2)  # custom grid
fig.tight_layout()
plt.show()
```
