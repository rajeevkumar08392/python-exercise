```python
d1={'data':[10,11,21,22,1,3,4,5,8,10]}
df_data=pd.DataFrame(d1)
df_data
```

```python
df_data['x-xmean']=df_data['data']-dmean
df_data
```

```python
varsq=sum(df_data['x-xmeansq'])/len(df_data)-1
varsq
```
