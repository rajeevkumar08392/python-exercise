```python
df_movie_title=pd.read_csv("/content/Movie_Id_Titles")
df_movie_title
```

```python
df_user_data=pd.read_csv("/content/u.data",sep='\t',names=["user_id","item_id","rating","timestamp"])
df_user_data.head()
```

```python
# merge()
df_movie_data=df_movie_title.merge(df_user_data,how="inner",on=["item_id"])
df_movie_data
```

```python
# sort_index()
# sort_Value()

# sort the data on bais of movie rating
# sort_value(columname,ascending=True) data will sorted in ascending order
df_sort_moviename=df_movie_data.sort_values("rating")
df_sort_moviename.head()
```

```python
#descending order
df_sort_moviename1=df_movie_data.sort_values("rating",ascending=False)
df_sort_moviename1.head()
```

```python
# to_csv(filename) Write object to a comma-separated values (csv) file.
df_sort_moviename1.to_csv("Moviesorteddata1.csv")
```

```python
d2={"productid":[100,101,103,105],"productcolor":['red','red','black','grey']}
df_inv=pd.DataFrame(d2)
df_inv
```

```python
#inner merge - it will take intersection of given column data
df_inner=pd.merge(df_product,df_inv,how="inner",on=["productid"])
df_inner
```

```python
# outer merge - it will take union of given column data
df_outer=pd.merge(df_product,df_inv,how="outer",on=["productid"])
df_outer
```

```python
# left merge - merge the data to left dataframe
df_left=pd.merge(df_product,df_inv,how="left",on=["productid"])
df_left
```

```python
df_right=pd.merge(df_product,df_inv,how="right",on=["productid"])
df_right
```
